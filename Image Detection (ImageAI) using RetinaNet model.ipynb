```python
#import the libraries
from imageai.Detection import ObjectDetection
import tensorflow as tf
import os
```

###Using RetinaNet Object detection

```python
#Run the Image detection model using RetinaNet object
execution_path = os.getcwd()

detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath( os.path.join(execution_path , "model/resnet50_coco_best_v2.0.1.h5"))
detector.loadModel()
```

```python
#Choose the type of objects that you want to detect
custom_objects = detector.CustomObjects(person=True,bicycle=False,car=False,motorcycle=False,airplane=False,bus=False,train=False,truck=False,bench=False,tie=True,suitcase=False,kite=False,bottle=False,cup=False,cake=False,chair=False,couch=False,potted_plant=False,bed=False,dining_table=True,tv=False,laptop=False,mouse=False,remote=False,keyboard=True,cell_phone=True,microwave=False,oven=False,toaster=False,sink=False,refrigerator=False,book=False,clock=False,vase=False,scissors=False,teddy_bear=False,toothbrush=False)
```

```python
#Run the dectetor model through the Input the image that you want to detect the objects 
detections = detector.detectCustomObjectsFromImage(input_image=os.path.join(execution_path , "Presidents.jpg"), output_image_path=os.path.join(execution_path , "Presidents_new.jpg"), custom_objects=custom_objects, minimum_percentage_probability=35)
```

```python
#iterate through all objects and print the name and probability in percentage
print("This is the reults of Image Detection model" "\n")

for eachObject in detections:
    print("-","{} is {}".format(eachObject["name"],eachObject["percentage_probability"].round(2)),"%" )
    
    
```
